```python
import networkx as nx 
  
edges = [(1, 2), (1, 6), (2, 3), (2, 4), (2, 6),(3, 4), (3, 5), (4, 8), (4, 9), (6, 7)] 

G = nx.Graph()
G.add_edges_from(edges) 
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold')
```

```python
options = {
    'node_color': 'black',
    'node_size': 100,
    'width': 3,
    'with_labels':True,
    'font_weight':'bold',
}
```

```python
nx.draw_shell(G, nlist=[range(5,10), range(5)],  **options)
```

```python
nx.draw_circular(G, **options)
```

```python
nx.draw_spectral(G, **options)
```

```python
nx.draw_random(G, **options)
```

```python
nx.shortest_path(G, 1, 3)
```

```python
nx.shortest_path(G, 1, 7)
```

```python
nx.shortest_path(G, 1, 9)
```

```python
nx.shortest_path(G, 1, 5)
```

# Multigraphs

```python
MG = nx.MultiGraph()
```

```python
MG.add_weighted_edges_from([(1, 2, 0.5), (1, 2, 0.75), (2, 3, 0.5)])
```

```python
MG.nodes()
```

```python
MG.add_weighted_edges_from([(3, 5, 0.5), (4, 6, 0.75), (4, 3, 0.5),(6, 8, 0.5), (7, 8, 0.75), (9, 1, 0.5),(8, 2, 0.5), (7, 3, 0.75), (10, 2, 0.5)])
```

```python
MG.nodes()
```

```python
MG.is_multigraph()
```

```python
MG.is_directed()
```

```python
MG.edges()
```

```python
for k,v in MG.adj.items():
    print("Edges untupled {0}, {1} ".format(k,v) )
```

```python
for k,v,w in G.edges.data('weight'):
    print("Edges untupled {0}, {1} and their corresponding weight{2} ".format(k,v,w) )
```

```python
for n, nbrs in MG.adjacency():
    for nbr, edict in nbrs.items():
        print(nbr,edict)
```

```python
nx.draw(MG, with_labels=True, font_weight='bold')
```

```python
nx.draw_circular(MG, **options)
```
